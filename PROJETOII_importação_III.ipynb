```python
origem = '/content/drive/MyDrive/Data/COMEXSTAT/'
```

```python
arquivo_1 = origem + 'IMP_2025_MUN.csv'
arquivo_2 = origem + 'IMP_2026_MUN.csv'
pais = origem + 'PAIS.csv'
uf_mun =origem + 'UF_MUN.csv'
ncm_sh = origem + 'NCM_SH.csv'
```

```python
imp25 = pd.read_csv(arquivo_1,low_memory=False,sep=';',encoding='UTF-8')
imp26 = pd.read_csv(arquivo_2,low_memory=False,sep=';',encoding='UTF-8')
imppais = pd.read_csv(pais,low_memory=False,sep=';',encoding='latin1')
impufmun = pd.read_csv(uf_mun,low_memory=False,sep=';',encoding='latin1')
impsh = pd.read_csv(ncm_sh,low_memory=False,sep=';',encoding='latin1')
```

```python
imp_final = pd.concat([imp25,imp26], ignore_index=True)
imp_final.info()
```

```python
imp_final.head(5)
```

```python
imp_final.tail(5)
```

```python
imp_final = imp_final.merge(impsh[['CO_SH4','NO_SH4_POR']], left_on='SH4', right_on='CO_SH4', how='left')
```

```python
imp_final = imp_final.merge(impufmun[['CO_MUN_GEO','NO_MUN_MIN']],left_on='CO_MUN', right_on='CO_MUN_GEO', how='left')
```

```python
imp_final = imp_final.merge(imppais[['CO_PAIS','NO_PAIS']],on='CO_PAIS',how='left')
```

```python
imp_final.info()
```

```python
imp_final.head(5)
```

```python
imp_final.tail(5)
```

```python
imp_final = imp_final.drop('SH4', axis=1)
```

```python
imp_final = imp_final.drop('CO_MUN_GEO', axis=1)
```

```python
imp_final.head(5)
```

```python
print(f"Shape of imp_final before dropping duplicates: {imp_final.shape}")
imp_final = imp_final.drop_duplicates()
print(f"Shape of imp_final after dropping duplicates: {imp_final.shape}")
```

```python
imp_final.to_csv(origem+'imp_final.csv',index=False)
```

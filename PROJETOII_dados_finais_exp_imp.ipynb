Cadeia Automotiva dados finais exportação e importação

```python
import pandas as pd
```

```python
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
origem = '/content/drive/MyDrive/Data/COMEXSTAT/'
```

```python
arquivo_1 = origem + 'dado_exp_final.csv'
arquivo_2 = origem + 'dado_imp_final.csv'
```

```python
expfinal = pd.read_csv(arquivo_1,low_memory=False,sep=',',encoding='UTF-8')
impfinal = pd.read_csv(arquivo_2,low_memory=False,sep=',',encoding='UTF-8')
```

```python
dado_final = pd.concat([expfinal,impfinal], ignore_index=True)
dado_final.info()
```

```python
dado_final.head(5)
```

```python
dado_final.tail(5)
```

```python
float_cols_to_convert = ['TOTAL KGL EXP.', 'TOTAL EXPORTADO', 'TOTAL KGL IMP.', 'TOTAL IMPORTADO']

for col in float_cols_to_convert:
    dado_final[col] = dado_final[col].fillna(0).astype(int)

print("Info after converting specified float columns to integer:")
dado_final.info()

print("\nHead of dado_final after conversion:")
print(dado_final.head(5))

print("\nTail of dado_final after conversion:")
print(dado_final.tail(5))

# Save the modified DataFrame
dado_final.to_csv(origem + 'dado_final.csv', index=False)
print(f"\nModified DataFrame saved to {origem + 'dado_final.csv'}")
```

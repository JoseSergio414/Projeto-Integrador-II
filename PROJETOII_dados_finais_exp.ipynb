```python
arquivo_1 = origem + 'exp_final_I.csv'
arquivo_2 = origem + 'exp_final_II.csv'
arquivo_3 = origem + 'exp_final_III.csv'
```

```python
exp21_22 = pd.read_csv(arquivo_1,low_memory=False,sep=',',encoding='UTF-8')
exp23_24 = pd.read_csv(arquivo_2,low_memory=False,sep=',',encoding='UTF-8')
exp25_26 = pd.read_csv(arquivo_3,low_memory=False,sep=',',encoding='UTF-8')
```
